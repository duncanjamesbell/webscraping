```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
```

```python
trump_urls = pd.read_csv('trump_urls.csv')
```

```python
trump_urls.head()
```

```python
url = "https://www.whitehouse.gov/briefings-statements/remarks-president-trump-thanksgiving-teleconference-members-military/"

html = urllib.urlopen(url).read()
```

```python
soup = BeautifulSoup(html, 'html.parser', from_encoding="utf-8")
```

```python
soup.find_all(name='p')
```

```python
url_text = []

for url in trump_urls.url:
    html = urllib.urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser', from_encoding="utf-8")
    url_text.append(soup.find_all(name='p'))
```

```python
with open('trump_text.txt', 'w') as f:
    for item in url_text:
        f.write("%s\n" % item)
```

```python
trump_urls['text'] = url_text
```
